```python
import cv2
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("D:/Bhoya/Tugas/Kuliah/Semester 4/PCD/Tugas 1/burano.jpg")
plt.imshow(img)
```

```python
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
```

```python
img_Gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_Gray, cmap="gray")
```

```python
img_HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img_HSV, cmap="hsv")
```

```python
img_HLS = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(img_HLS)
```

```python
gmbr = cv2.imread("D:/Bhoya/Tugas/Kuliah/Semester 4/PCD/Tugas 1/tcinta.jpg")
gmbr_copy = img.copy()
cv2.rectangle(gmbr, pt1 = (800,470), pt2 = (980, 530), color = (255, 0, 0), thickness = 5)
plt.imshow(gmbr)
```

```python
cv2.circle(gmbr, center = (950, 50), radius = 50, color = (0, 0, 255), thickness = 5)
plt.imshow(gmbr)
```

```python
cv2.putText(gmbr, text = "the Wall of Love", org = (250,250), fontFace = cv2.FONT_HERSHEY_DUPLEX, fontScale = 2, color = (0, 255, 0), thickness = 2, lineType = cv2.LINE_AA)
plt.imshow(gmbr)
```
